```python
import pandas as pd
df = pd.read_csv("external_data.csv")
```

```python
df.info()
```

```python
df["date"]
```

```python
df['DateTime'] = pd.to_datetime(df["date"], dayfirst=True)
```

```python
df["lockdown"] = ((df['DateTime'] >= '2020-03-16') & (df['DateTime'] <= '2020-05-11')
| ((df['DateTime'] >= '2020-10-28') & (df['DateTime'] <= '2020-12-15'))
| ((df['DateTime'] >= '2021-04-28') & (df['DateTime'] <= '2021-05-03')))
```

```python
df["lockdown"].sum()
```

```python
df['hour'] = df['DateTime'].dt.hour
```

```python
df["curfew"] = (((df['DateTime'] >= '2021-05-03') & (df['DateTime'] <= '2021-05-18') 
                & (df['DateTime'].dt.hour >= 19) | (df['DateTime'].dt.hour <= 6))
                | ((df['DateTime'] >= '2021-05-19') & (df['DateTime'] <= '2021-06-08') 
                & (df['DateTime'].dt.hour >= 21) | (df['DateTime'].dt.hour <= 6))
                | ((df['DateTime'] >= '2021-06-09') & (df['DateTime'] <= '2021-06-29')
                & (df['DateTime'].dt.hour >= 23) | (df['DateTime'].dt.hour <= 6)))
```

```python
df["curfew"].sum()
```
